```python
def difference_histogram(x,y, bins=32, bin_range=(1,40)):    from scipy.spatial.distance import pdist    # Calculate all distances between all (x,y) pairs    distances = []    vec = np.zeros((len(x),2))    vec[:,0] = x    vec[:,1] = y        # Compute pair-wise distances (we're lucky scipy has a built-in for this ...)    distances = pdist(vec)        # Now, take a histogram of those pair-wise distances. That's our representation.    return np.histogram(distances, bins=bins, range=bin_range)[0]
```

```python
def show_dendrogram_and_mountain_plot(state_probabilities, sub_state_mapping, Z=None, cophenetic_distance=4.0, colormap=cm.bone, smooth_std=30):    """    state_probabilities - (numFrames x numStates) array, indicating probability of being in one state or another at a given point in time.    states - (numFrames x 1) array, indicating the chosen state at a givne frame.    sub_state_mapping - (numStates) array. Each state has been clustered previously to group together into a new state. This array describes that mapping.    Z - we like dendrograms. This holds the linkage information from a call to scipy.cluster.hierarchy.linkage()    cophenetic_distance - this is a threshold for how to color the dendrogram.    colormap - the colormap used in the mountain plot    smooth_std - the timescale over which to smooth the mountain data.    """        smoothed_prob = qbp.smooth_channels(state_probabilities.copy(), smooth_std)        figure(figsize=(20,5))    subplot(1,2,1, position=(0,1,0.15,1.0))    R = hierarchy.dendrogram(Z, orientation='right', labels=new_identities, count_sort=True)    new_order = R['leaves'] # reorder the state probabilities so the dendrogram matches up with the flow plot    yl = gca().get_ylim()#    axis('off')        num_clusters = len(np.unique(sub_state_mapping))    lookup = np.choose(np.sort(sub_state_mapping), np.linspace(0,0.8,num_clusters))    colors = cm.bone(lookup)    smoothed_prob = qbp.smooth_channels(state_probabilities.copy(), smooth_std)    smoothed_prob = smoothed_prob[:,np.argsort(sub_state_mapping)[::-1]]        subplot(1,2,2, position=(0.16,1, 0.65, 1.0))    qbp.stacked_plot(smoothed_prob, normalize=True, colors=colors, colormap=colormap)        return
```
