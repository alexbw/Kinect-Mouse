```python
def nparrays_from_frames(frames):    arrays = []    for key in frames[0].keys():        if key[0] == '_':            continue        if 'distance' in key:            data = np.asarray( [f[key] for f in frames] )            data /= data.max()            arrays.append(data)        else:            arrays.append( np.asarray( [f[key] for f in frames] )  )            if arrays[-1].ndim == 1:            arrays[-1] = np.atleast_2d(arrays[-1]).T            return np.hstack(arrays)
```

```python
def get_array_from_keys(frames, some_key_set):    arrays = []    for key in some_key_set:        arrays.append( np.asarray( [f[key] for f in frames] )  )            if arrays[-1].ndim == 1:            arrays[-1] = np.atleast_2d(arrays[-1]).T        return np.hstack(arrays)def calc_cluster_silhouette(p_data, num_clusters=12, n_tries=100, silhouette_sample_size=1000):    # Cluster it    km = MiniBatchKMeans(k=num_clusters, init='k-means++', compute_labels=True)    km.fit(p_data)        silhouette = []    for i in range(n_tries):        sil = silhouette_score(p_data, km.labels_, sample_size=silhouette_sample_size, metric='euclidean')        silhouette.append( sil )            return np.asarray(silhouette)def identically_distributed_noise(some_array):    noise_array = np.zeros_like(some_array)    for i in range(some_array.shape[1]):        noise_array[:,i] = np.random.normal( np.mean(some_array[:,i]), np.std(some_array[:,i]), size=some_array[:,i].shape)    return noise_array
```
