```python
def difference_histogram(x,y, bins=32, bin_range=(1,40)):    from scipy.spatial.distance import pdist    # Calculate all distances between all (x,y) pairs    distances = []    vec = np.zeros((len(x),2))    vec[:,0] = x    vec[:,1] = y        # Compute pair-wise distances (we're lucky scipy has a built-in for this ...)    distances = pdist(vec)        # Now, take a histogram of those pair-wise distances. That's our representation.    return np.histogram(distances, bins=bins, range=bin_range)[0]
```
