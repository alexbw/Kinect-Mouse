```python
def full_diff(someArray, the_axis=0):    output = np.zeros_like(someArray)    output[:-1] = np.diff(someArray, axis=the_axis)    return outputdef velocity_from_positions(x,y):    x = np.asarray(x)    y = np.asarray(y)    distances = np.zeros_like(x)    distances[:-1] = np.sqrt( (x[1:] - x[:-1])**2 + (y[1:] - y[:-1])**2 )    return distancesdef distance_from_point(x,y,source_x, source_y):    """    Find euclidean distance from some (x,y) points (x and y are both 1D arrays)    to some source point (source_x, source_y).     e.g. the distance of a mouse's head to an odor source.    """    x = np.asarray(x)    y = np.asarray(y)    return np.sqrt( (x - source_x)**2 + (y - source_y)**2 )def get_quadrant_location(x,y,half_x,half_y):    quadrant_locations = np.zeros_like(x)        idx_left = x <= np.asarray(half_x)    idx_bottom = y <= np.asarray(half_y)        quadrant_locations[ idx_left & ~idx_bottom ] = 1    quadrant_locations[ ~idx_left & ~idx_bottom ] = 2    quadrant_locations[ idx_left & idx_bottom ] = 3    quadrant_locations[ ~idx_left & idx_bottom ] = 4        return quadrant_locations
```
