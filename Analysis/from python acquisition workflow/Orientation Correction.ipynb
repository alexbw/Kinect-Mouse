Now that we've gotten all the spline fits and images, we'll need to rotate the images so that the mouse is always facing right(this is "right", as defined by how the mouse was pulled out. This is not always correct, but we'll be dealing with that later on).

```python
def rotateAndCrop(img, angle, origSize=(80,40), downSize=(40, 20)):	import Image	rotatedImg = Image.fromarray(img).rotate(-angle, expand=True)		# Crop the PIL image to a standard frame, then downsize it	xCenter, yCenter = rotatedImg.size	xCenter /= 2	yCenter /= 2	cropRegion = (xCenter-origSize[0]/2, 				  yCenter-origSize[1]/2, 				  xCenter+origSize[0]/2, 				  yCenter+origSize[1]/2)	rotatedImg = rotatedImg.crop(cropRegion).resize(downSize)	return np.asarray(rotatedImg)def correctImages(imgs, diffimgs):    for i in range(len(diffimgs)):        angle = ellipses[i][0][2]            imgs[i][0]      = rotateAndCrop(imgs[i][0], angle)        diffimgs[i][0]  = rotateAndCrop(diffimgs[i][0], angle)    return imgs, diffimgs
```

## 3. Orientation correction.    This is hard. But, the general strategy is this:  #### For each frame,    - For the difference image, split it in half. Sum each half separately.    - Calculate the difference between the two halves. Save it.  #### When differences are calculated, for a sliding window calculated over every frame in temporal order,     - Calculate the skewness and significance of the skew for the distribution made with every frame in the window#### When skewness is calculated,  - Find blocks of frames that contain five consecutive singificantly incorrectly (left) skewed frames  - Mark those frames to be flipped later  - Find blocks of frames that contain a flip sandwiched between five consecutive significantly correctly (right) skewed frames to significantly incorrectly (left) skewed frames  - Mark those frames to be thrown out#### When the frame blocks for flipping and removal are found  - Flip the frames that need flipping  - Remove the frames that need removing

```python
# Calculate the left-right difference datadef getMovementDiffs(diffMouseImages):    theMiddle = diffMouseImages[0][0].shape[1] / 2    diffs = [(np.sum(d[0][:,:theMiddle]), np.sum(d[0][:,theMiddle:])) for d in diffMouseImages]    leftDiffs = np.asarray([d[0] for d in diffs])    rightDiffs = np.asarray([d[1] for d in diffs])    movementDiff = rightDiffs - leftDiffs # this is the key dataset here        return movementDiff
```
