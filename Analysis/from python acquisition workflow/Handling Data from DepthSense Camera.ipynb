```python
def densify(img, windowsize=7):	windowSize = [windowsize, windowsize]	img = img.astype('float32')	img[np.isnan(img)] = -np.inf	while np.isinf(img).sum() > 0:		idx = np.isinf(img)		for i in range(windowSize[0]):			for j in range(windowSize[1]):				shifted = np.roll(np.roll(img,-windowSize[0]/2+i,0),-windowSize[1]/2+j,1)				img[idx] = np.maximum(img[idx], shifted[idx])	return img
```
