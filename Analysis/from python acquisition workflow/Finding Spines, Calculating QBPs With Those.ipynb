Use a more processed dataset for QBP calculation. I've got these wonderful Kalman-filtered spline fits, and it's silly to not try using them.  This notebook tests the second approach.

### Let's find the head and the tail.    The tail we've already found previously, when calculating the spline fit. The head is a little trickier, because it's not always halfway around the spline. For instance,  when the animal is grooming himself , his head is placed by his side and pointed down and away.

```python
def findHead(fit):    """    In the particular format we use for spline fits,     it's shaped like (2,n) (where n is usually 50),    and the first entry, fit[:,0] is the butt location.        Using this format, we can find the head as a point of high curvature    at some distance away from the butt.    """    numPointsIn = fit.shape[1]    tck, u  = interp.splprep(fit, k = 4, s=0.5, per=1)    u = np.linspace(0,1,numPointsIn)    x,y = interp.splev(u, tck)        ddx,ddy = interp.splev(u, tck, 2)    curvature = ddx**2 + ddy**2    head_search_range = np.r_[ numPointsIn/4 : 3*numPointsIn/4 ]    head_idx = head_search_range[0] + np.argmax(curvature[head_search_range])        return head_idx, curvature
```
