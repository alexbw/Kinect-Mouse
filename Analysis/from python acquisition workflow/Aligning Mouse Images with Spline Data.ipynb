### The plan:First, we've got to transform the spline fits into the image space. This means:   - Subtract the mean of the spline   - Rotate it to face right   - Scale down by 4 (or two, I forget the scale factor that creates the images)   - Find the offset from the current butt-position to a standardized butt-position   - Shift the image and difference image but this offset.   - Buy 3 grams of pure cocaine from a guy named Jean-Luc who keeps his shirt half-open all the time   - Turn off my phone and seek enlightenment  

```python
scaleFactor = 2buttPosition = 5 # all butts will be placed at this pixel x locationdef rotate(xyPoints, angle):    """    xyPoints - 2 x n array of x,y points to be rotated around 0    """    angle = ( angle / 180 ) * np.pi    rotationMatrix = np.r_[ [[np.cos(angle), -np.sin(angle)], [np.sin(angle), np.cos(angle)]] ]    return np.dot( rotationMatrix, xyPoints )def realignMeanCenteredImagesToRearEnd(fits, ellipses, imgs, diffimgs):    numFrames = len(fits)        allOffsets = [0]*numFrames    allFits = [0]*numFrames    allImgs = [0]*numFrames    allDiffImgs = [0]*numFrames    for i in range(numFrames):        angle = ellipses[i][0][2]                thisFit = np.copy(fits[i])[0,:,:]        thisFit[0,:] -= thisFit[0,:].mean() # x        thisFit[1,:] -= thisFit[1,:].mean() # y                thisFit = rotate(thisFit, -angle)        thisFit /= scaleFactor                allFits[i] = thisFit                # Find the point on the y=0 line with the smallest x        left_side = np.where(thisFit[1,:] < (imgs[0].shape[1]/2.5 - imgs[0].shape[1]/2))[0]        butt_idx = np.argmin(np.abs(thisFit[0,left_side]))        butt_idx = left_side[butt_idx]        currentButtPosition = imgs[0].shape[1]/2 + thisFit[1,butt_idx]        offset = np.floor( buttPosition - currentButtPosition ).astype('int')                allOffsets[i] = offset                allImgs[i] = np.roll(imgs[i], offset, 1)        allDiffImgs[i] = np.roll(diffimgs[i], offset, 1)            return allFits, allImgs, allDiffImgs, allOffsets
```

### Saving the Results Out

```python
fits1, imgs1, diffimgs1, _ = realignMeanCenteredImagesToRearEnd(fits1, ellipses1, imgs1, diffimgs1)fits2, imgs2, diffimgs2, _ = realignMeanCenteredImagesToRearEnd(fits2, ellipses2, imgs2, diffimgs2)
```

```python
outfile1 = "processed_mice/10-24 001 blank shifted aligned.npz"outfile2 = "processed_mice/10-24 003 blank shifted aligned.npz"np.savez(outfile1, fits=fits1, ellipses=ellipses1, imgs=imgs1, diffimgs=diffimgs1, frame_ids=frame_ids1)np.savez(outfile2, fits=fits2, ellipses=ellipses2, imgs=imgs2, diffimgs=diffimgs2, frame_ids=frame_ids2)
```
